Merging Datasets

```python
# merging studentVle and Vle
combined_vle = pd.merge(studentVle_df2, vle_df2, on = "id_site", how = 'inner', suffixes=("", "_vle"))
combined_vle
```

```python
# dropping duplicates  and less relevant columns from merged df
combined_vle2 = combined_vle.drop(['code_mod_pres_vle', 'id_site', 'week_from', 'week_to', 'date'], axis=1)
combined_vle2
```

```python
# merging studentInfo Vle
stu_info_vle= pd.merge(studentInfo_df2, combined_vle2, on = "unique_identifier", how = 'outer', suffixes=("", "_vle"))
# dropping duplicates  and less relevant columns from merged df
stu_info_vle2 = stu_info_vle.drop(['id_student_vle', 'code_mod_pres_vle'], axis=1)

stu_info_vle2
```

```python
# merging assessments and studentAssessments
combined_assessments= pd.merge(studentAssessment_df, assessments_df2, on = "id_assessment", how = 'inner')

# dropping less relevant columns
combined_assessments_2 = combined_assessments.drop(['date', 'weight', 'date_submitted', 'is_banked', 'code_mod_pres', 'id_assessment'], axis=1)

combined_assessments_2
```

```python
# merging combined_assessments and stu_info_vle
combined_grade= pd.merge(stu_info_vle2, combined_assessments_2, on = "id_student", how = 'inner')
combined_grade
```

```python
#saving to merged data to csv
combined_grade.to_csv("merged_data.csv")
```
